First we create a simple world consisting of two joints, where the first joint can unlock the other joint at 90 +/- 2 degrees.

```python
def create_world():
    noise = { 'q' : .5, 'vel' : .4}
    joints = []
    
    states = [88, 92]
    dampings = [15, 200, 15]
    limits = [0, 180]
    joints.append(Joint(states, dampings, limits, noise))
    
    states = [90]
    dampings = [2, 2]
    limits = [ 0, 90 ]
    joints.append(Joint(states, dampings, limits, noise))
    
    world = World(joints)
    
    locker1 = Locker(world, locker=world.joints[0], locked=world.joints[1], lower=0, upper=85)
    locker2 = Locker(world, locker=world.joints[0], locked=world.joints[1], lower=95, upper=180)
    
    return world
```

```python
world = create_world()
controller = Controller(world)
recorder = Recorder(world)
```

Now we command the controller to move the first joint to 120 degree. This is achieved by using a simple PID controller, that applies forces to the joint directly.

```python
controller.move_to(0, 120)
while not controller.done:
    world.step(.1)
```

Given this movement of the joint we want to find potential 'changepoints' in the world, that could be a hint of something interesting - e.g. a new unlocked joint. This is done by Bayesian changepoint detection as described in Paul Fearnhead, Exact and Efficient Bayesian Inference for Multiple
Changepoint problems, Statistics and computing 16.2 (2006), pp. 203--213. As input data we use the applied force to the joint. This simulates our force-torque sensor in the PR2's wrist.

```python
import bayesian_changepoint_detection.offline_changepoint_detection as bcd
from functools import partial
```

Since our model assumens piecewise constant data (+ gaussian noise), but the force we apply is roughly piecewise linear, we compute the first differential.

```python
df = np.array(controller.desired_forces)
ddf = np.ndarray(len(df)-1)
for i in range(len(ddf)):
    ddf[i] = df[i+1] - df[i]
```

```python
Q, P, Pcp = bcd.offline_changepoint_detection(data=ddf,
                                              prior_func=partial(bcd.const_prior, l=(len(ddf)+1)), 
                                              observation_log_likelihood_function=bcd.gaussian_obs_log_likelihood)
```

```python
fig, ax = plt.subplots(figsize=[18, 16])
ax = fig.add_subplot(3, 1, 1)
ax.plot(df[:])
ax = fig.add_subplot(3, 1, 2, sharex=ax)
ax.plot(ddf[:])
ax = fig.add_subplot(3, 1, 3, sharex=ax)
ax.plot(np.exp(Pcp).sum(0))
```

We transfer the changepoint probabilities from the time domain to the joint space. We therefore check at which position we measured the joint to be at at each timeslice and compute the mean probability $f(x)$ for each position in joint space.

```python
P = np.exp(Pcp).sum(0)
print(len(P))
```

```python
def get_probability_over_degree(P, recorder):
    probs = np.zeros((360,))
    count = np.zeros((360,))
    for i, pos in enumerate(recorder.qs):
        if i < 2:
            continue

        deg = int(pos[0])%360
        
        probs[deg] += P[i-3]
        count[deg] += 1

    probs = probs/count
    prior = 10e-15
    probs = [prior if np.isnan(p) else p for p in probs]
    return probs
```

```python
probs = get_probability_over_degree(P, recorder)
plt.plot(probs)
```

As next step we need to sample actions and for each sampled action we sample from the distribution of joint parameters $P(\theta)$. We then simulate the action with the given $\theta$ and get samples from the distribution of endpositions of the joint. In 3D we would like to project that back to 3D space given the sample of $\theta$ in this simplified version we don't need to.

An action we can sample is a starting point for both joints and an desired force over time applied of one of the joints.

```python
def sample_action(probs):
    return {'positions': {'0': np.random.randint(0, 180), 
                          '1': np.random.randint(0, 90)}, 
            'goal': {'force': np.random.uniform(-80, 80), 
                     'time': np.random.randint(1000, 10000)}
            }
```

We also want to sample the world state, given the observations. Thus we use the change point distribution $f(x)$ to generate samples using "Gibbs" sampling. This is somewhat overexaggerated, since we assume the samples at each point to be independent and thus the Gibbs sampling becomes utterly easy.

```python
def f(x, probs):
    return np.prod([probs[int(p)] for p in x])
```

```python
def gibbs_proposal(probs):
    x = []
    for i, p in enumerate(probs):
        if np.random.uniform() < p:
            x.append(i)
    return x
```

```python
for i in range(30):
    print(gibbs_proposal(probs))
```

We now have to compute the posterior over the friction constant at each part. So what we could do is to cut the observations in pieces, where they just belong to each part and then run the MCMC method from our IROS paper. 

But for the sake of easy implementation we just ignore the friction for the moment and simply set them to the correct value. We also ignore that the limits are normally must be inferred as well and just set them to the right values.

Our agent also assumes that the second joint is locked at the beginning and that every change point from the first joint might change that with probability $p$.

```python
def sample_lock(cps, p):
    locks = []
    for cp in cps:
        if np.random.uniform() < p:
            locks.append(cp)
    return locks
```

```python
locks = sample_lock(gibbs_proposal(probs), 0.5)
print(locks)
```

Now build up a world and run the simulation until nothing moves anymore. In real world this might be PhysX or ODE, here we have a much simpler forward model we use. We then get samples from trajectories of the joints. We use the endpositions as our samples and compute the entropy of them.

```python
def create_sim_world(locks_one, locks_two):
    noise = { 'q' : .5, 'vel' : .4}
    joints = []
    
    states = [88, 92]
    dampings = [15, 200, 15]
    limits = [0, 180]
    joints.append(Joint(states, dampings, limits, noise))
    
    states = [90]
    dampings = [2, 2]
    limits = [ 0, 90 ]
    joints.append(Joint(states, dampings, limits, noise))
    
    world = World(joints)
    
    lock_size = 1
    locks = [0]
    for lock in locks_one:
        locks.append(lock-lock_size)
        locks.append(lock+lock_size)
    locks.append(360)
    
    lock_tuples = []
    for i in xrange(0, len(locks), 2):
        lock_tuples.append((locks[i], locks[i+1]))
    
    MultiLocker(world, locker=world.joints[0], locked=world.joints[1], locks=lock_tuples)
    
    return world
```

Ok, now everything is in place to generate endposition samples. Let's generate some.

```python
def endpos_samples(probs, Na, Ntheta):
    lock_dist = np.zeros((180,))
    distribution = [[] for i in range(360)]
    for i in range(Na):
        start_pos = int(np.random.uniform(0, 180))
        end_pos = []
        for j in range(Ntheta):
            action = sample_action(probs)
            cps = gibbs_proposal(probs)
            locks = sample_lock(cps, 0.9)
            for l in locks:
                lock_dist[l] += 1
            sim_world = create_sim_world(locks, [])
            sim_recorder = Recorder(sim_world)
            fc = ForceController(sim_world, sim_world.joints[1], action['goal']['time'], action['goal']['force'])
            
            sim_world.joints[0].q = start_pos
            sim_world.joints[1].q = 0
            sim_world.joints[0].vel = 0
            sim_world.joints[1].vel = 0
            
            # actually run simulation ('forward model')
            for t in range(int(fc.time)):
                sim_world.step(.1)
            
            while sim_world.joints[1].vel > 10e-3:
                sim_world.step(.1)
            
            end_pos.append(sim_world.joints[1].q)
            distribution[start_pos].append(sim_world.joints[1].q)
    return (distribution, lock_dist)
```

```python
data, lock_dist = endpos_samples(probs, 10000, 1)
```

Now let's plot the endposition distribution samples. And also some other metrics, like the standard deviation (proportional to the entropy) and a Gaussian match to mean and standard deviation. (Maybe I should use pandas here...)

```python
def plot_std(data):
    fig, ax = plt.subplots(figsize=[20, 15])
    stdarr = np.ndarray((360,))
    for p, d in enumerate(data):
        stdarr[p] = np.std(d)
    ax.plot(stdarr)
    return np.array(stdarr)

def plot_gauss(data):
    fig, ax = plt.subplots(figsize=[20, 15])
    for p, d in enumerate(data):
        if len(d) > 0:
            ax.plot([p]*2, [np.mean(d)-np.std(d)/np.sqrt(len(d)), np.mean(d)+np.std(d)/np.sqrt(len(d))])
            ax.plot([p], [np.mean(d)], '.')
            
def plot_data(data):
    fig, ax = plt.subplots(figsize=[20, 15])
    for p, d in enumerate(data):
        if len(d) > 0:
            ax.plot([p]*len(d), d, 'o')
    
```

```python
plot_data(data)
```

```python
plot_std(data)
```

```python
plot_gauss(data)
```

As one can see, the three distributions are very similar -- and what is most important have their maxima at the exact same position. So for the easy case we know, that the position of maximum entropy is the position of the highest likelihood of changepoints! Thus we can save a lot of computational cost, by directly using the changepoint distribution.

However, if we have modelled different joints already, this picture changes slightly. First we have to model, which joint locks which other joint. But this is important anyway. The second thing is, that now we have to integrate the knowledge about other joints in our model, s.t. the endposition distribution will reflect how much we know about the joint. For that purpose, we could fix the amount of force we apply to a joint. Then we would only sample from $P(\theta)$ (the distribution over joint parameters) and not really actions or we middle the entropy of the different enposition distributions over different forces. However we solve this, we still can infer this independently from the locking state and then multiply with the probability of being locked, which somehow is proportional to the changepoint probability of the locking joint. This greatly saves compute power.
